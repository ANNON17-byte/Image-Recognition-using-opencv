```python
img = cv2.imread("/content/drive/MyDrive/napo.png")
img.shape # (height, width , channel)
```

```python
img = cv2.resize(img, (200,200))
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # In OpenCV , BGR format is followed instead of RGB
cv2_imshow(gray) #gray has only 1 channel
```

```python
#if we want to change the channel of image but not its color, then we use the hsv formaat
#HSV - hue, saturation , variance(Brightness)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)
```
